# Filtering

```python
import glob, os
from obspy import read, Stream
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Get list of the folders


```python
cwd = os.getcwd()

dirs = [ name for name in os.listdir(os.getcwd()) if os.path.isdir(os.path.join(os.getcwd(), name)) ]
```

### Remove unnecessary folder

```python
dirs.remove('.ipynb_checkpoints')
```

### Get the list of the files inside in a folder

```python
sacs = glob.glob(dirs[0] + '/*')
```

## Read all sac files 

```python
st_org = Stream()
for sac in sacs:
    st_org += read(sac,format='SAC')
```

```python
## Filter the data
```

```python
st_lowpass = st_org.copy()
st_highpass = st_org.copy()
st_bandpass = st_org.copy()

st_highpass.filter('highpass',freq = 3)
st_highpass.plot()
```

```python
st_lowpass.filter('lowpass',freq = 1)
st_lowpass.plot()
```

```python
st_bandpass.filter("bandpass", freqmin=1, freqmax=2)
st_bandpass.plot()
```

#### Filter only one of the trace (3rd one)

```python
st_bandpass[2].filter("bandpass", freqmin=1, freqmax=2)
st_bandpass[2].plot()
```

#### Filter the traces one by one 

```python
for tr in st_bandpass:
    tr.filter('highpass',freq = 3)
    tr.plot() 
```
